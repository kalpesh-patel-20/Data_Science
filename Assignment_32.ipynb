### **Q1: Difference between a t-test and a z-test**

| Feature      | Z-Test                            | T-Test                            |
|--------------|------------------------------------|------------------------------------|
| Use case     | Population standard deviation known | Population standard deviation unknown |
| Sample size  | Typically n > 30                   | Typically n < 30                   |
| Distribution | Standard Normal (Z)                | Student’s T                        |

**Example:**
- **Z-test**: Testing if a factory’s average battery life is 1000 hrs when σ = 50 is known.
- **T-test**: Testing average exam score of 25 students when σ is unknown.

### **Q2: One-tailed vs. Two-tailed Tests**

| Test Type    | Description                                      | Hypotheses Example                 |
|--------------|--------------------------------------------------|------------------------------------|
| One-tailed   | Tests if a parameter is > or < a value           | H₀: μ = 50, H₁: μ > 50             |
| Two-tailed   | Tests if a parameter is ≠ a value                | H₀: μ = 50, H₁: μ ≠ 50             |

---

### **Q3: Type I & Type II Errors**

- **Type I Error (False Positive)**: Rejecting H₀ when it's true.
  - *Example*: Concluding a drug is effective when it’s not.
- **Type II Error (False Negative)**: Failing to reject H₀ when it's false.
  - *Example*: Concluding a drug is ineffective when it is effective.

### **Q4: Bayes’ Theorem (Concept + Example)**

**Formula**:
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

**Example**:
- Disease prevalence (A): 1% (0.01)
- Test sensitivity (P(B|A)): 99%
- False positive rate (P(B|¬A)): 5%

\[
P(A|B) = \frac{0.99 \cdot 0.01}{0.99 \cdot 0.01 + 0.05 \cdot 0.99} \approx 0.166
\]

So, only ~16.6% chance the person has the disease even with a positive test.

### **Q5: Confidence Interval (CI)**

CI gives a range of values for a population parameter.

**Formula**:
\[
CI = \bar{x} \pm z \cdot \frac{s}{\sqrt{n}}
\]

**Example**:
- Mean = 70, SD = 10, n = 25, Confidence = 95%
\[
CI = 70 \pm 1.96 \cdot \frac{10}{\sqrt{25}} = 70 \pm 3.92 = (66.08, 73.92)
\]

### **Q6: Bayes’ Theorem Sample Problem**

**Problem**:
- 2% of packages are defective (P(D) = 0.02)
- Inspection test detects 95% of defective (P(Pos|D) = 0.95)
- 10% false positive rate (P(Pos|¬D) = 0.10)

\[
P(D|Pos) = \frac{0.95 \cdot 0.02}{0.95 \cdot 0.02 + 0.10 \cdot 0.98} \approx 0.162
\]

### **Q8: Margin of Error & Sample Size Effect**

**Formula**:
\[
\text{Margin of Error} = z \cdot \frac{s}{\sqrt{n}}
\]

- **Larger sample size → smaller margin of error**
- *Example*: For a survey:
  - n = 100 → MoE ≈ ±3%
  - n = 1000 → MoE ≈ ±1%

### **Q9: Z-score Calculation**

\[
z = \frac{X - \mu}{\sigma} = \frac{75 - 70}{5} = 1.0
\]

**Interpretation**: The data point is 1 standard deviation above the mean.

### **Q10: Weight Loss Drug Effectiveness (t-test)**

Given: x̄ = 6, s = 2.5, n = 50, μ₀ = 0  
\[
t = \frac{6 - 0}{2.5 / \sqrt{50}} = \frac{6}{0.3536} ≈ 16.98
\]

Compare with critical t (df = 49, α = 0.05): t_critical ≈ 2.01  
Since 16.98 >> 2.01 → **Reject H₀**, drug is effective.

---

### **Q11: CI for Proportion (p = 0.65, n = 500)**

\[
SE = \sqrt{\frac{p(1-p)}{n}} = \sqrt{\frac{0.65 \cdot 0.35}{500}} \approx 0.0216
\]
\[
CI = 0.65 \pm 1.96 \cdot 0.0216 = (0.608, 0.692)
\]

**Interpretation**: 60.8% to 69.2% of population likely satisfied.

### **Q12: Compare Two Teaching Methods (t-test)**

Sample A: x̄₁ = 85, s₁ = 6, n₁ = 30  
Sample B: x̄₂ = 82, s₂ = 5, n₂ = 30

\[
SE = \sqrt{\frac{6^2}{30} + \frac{5^2}{30}} = \sqrt{1.2 + 0.833} = \sqrt{2.033} ≈ 1.426
\]
\[
t = \frac{85 - 82}{1.426} ≈ 2.10
\]

df ≈ 58, t_critical (α = 0.01, two-tailed) ≈ 2.66 → **Fail to reject H₀**

### **Q13: CI with Population Mean = 60, SD = 8, Sample Mean = 65, n = 50, Confidence = 90%**

\[
z = 1.645, \quad SE = \frac{8}{\sqrt{50}} ≈ 1.131
\]
\[
CI = 65 \pm 1.645 \cdot 1.131 = 65 \pm 1.861 = (63.14, 66.86)
\]

### **Q14: Caffeine Reaction Time (t-test)**

x̄ = 0.25, μ₀ = (assumed) 0.3, s = 0.05, n = 30

\[
t = \frac{0.25 - 0.3}{0.05 / \sqrt{30}} = \frac{-0.05}{0.0091} ≈ -5.48
\]

t_critical (α = 0.10, one-tailed, df=29) ≈ 1.31  
Since -5.48 < -1.31 → **Reject H₀**, caffeine **significantly** affects reaction time.
