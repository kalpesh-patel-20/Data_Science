### **Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)?**

- **PMF (Probability Mass Function)** is used for **discrete random variables**. It gives the probability that a discrete random variable is **exactly equal** to a specific value.

  **Example:** Rolling a die:  
  \( P(X = 4) = \frac{1}{6} \)

- **PDF (Probability Density Function)** is used for **continuous random variables**. It represents the **relative likelihood** for this random variable to take on a given value. The probability for a specific point is **zero**, but we can find the probability **within a range** by integrating the PDF.

  **Example:** Height of people (continuous variable):  
  Probability that height lies between 160 cm and 170 cm is given by the **area under the PDF curve** from 160 to 170.

---

### **Q2: What is Cumulative Density Function (CDF)? Why is it used?**

- The **Cumulative Density Function (CDF)** gives the **probability that a random variable X is less than or equal to a particular value x**, i.e.  
  \[ CDF(x) = P(X \leq x) \]

- **Example:** In a standard normal distribution,  
  \( P(X \leq 1.96) = 0.975 \) → This is from the CDF table.

- **Why use CDF?**  
  - To compute probabilities for intervals.  
  - To compare distributions.  
  - It's used in hypothesis testing and confidence intervals.

### **Q3: Where is Normal Distribution Used? What affects its shape?**

**Use cases of normal distribution:**
- Heights of people
- Test scores (like SAT, IQ)
- Measurement errors
- Blood pressure readings

**Parameters:**
- **Mean (μ)** → Determines the **center** of the curve.
- **Standard Deviation (σ)** → Determines the **spread/width**:
  - Larger σ → Flatter and wider curve.
  - Smaller σ → Taller and narrower curve.

---

### **Q4: Importance of Normal Distribution**

- Many natural phenomena follow normal distribution.
- It’s the foundation for many statistical tests and machine learning algorithms.
- Used in:
  - Quality control
  - Finance (returns modeling)
  - Health (blood pressure, BMI distribution)

### **Q5: Bernoulli Distribution vs Binomial Distribution**

- **Bernoulli Distribution:**  
  Only **one trial** with two outcomes (success = 1, failure = 0).  
  **Example:** Tossing a coin once.

- **Binomial Distribution:**  
  A **series of independent Bernoulli trials** (n trials).  
  **Example:** Tossing a coin 10 times and counting heads.

| Feature | Bernoulli | Binomial |
|--------|-----------|----------|
| Trials | 1         | n        |
| Output | 0 or 1    | 0 to n   |
| Parameter | p (prob. of success) | n (trials), p (success prob) |

### **Q6: Probability X > 60 (Normal Distribution)**

Given:
- Mean (μ) = 50
- Standard Deviation (σ) = 10
- X = 60

Step 1: Calculate Z-score  
\[
Z = \frac{X - \mu}{\sigma} = \frac{60 - 50}{10} = 1.0
\]

Step 2: Use Z-table to find \( P(Z < 1.0) = 0.8413 \)

Step 3:  
\[
P(X > 60) = 1 - 0.8413 = 0.1587
\]

✅ **Answer:** There's a **15.87%** chance that a randomly selected value is greater than 60.

### **Q7: What is Uniform Distribution?**

A **Uniform Distribution** is a probability distribution where **all outcomes are equally likely**.

- **Discrete Example:** Rolling a fair die (each side has 1/6 probability)
- **Continuous Example:** Random number generator between 0 and 1, where every value has equal probability density.

---

### **Q8: What is Z-score? Why is it important?**

- **Z-score** measures how many **standard deviations** an element is from the mean:
  \[
  Z = \frac{X - \mu}{\sigma}
  \]

**Importance:**
- Helps identify **outliers**.
- Standardizes data for comparison.
- Used in hypothesis testing and confidence intervals.

---

### **Q9: What is the Central Limit Theorem (CLT)? Why is it important?**

- **CLT:** The distribution of the sample mean of **any independent, random variable** becomes **approximately normal** if the **sample size is large enough**, regardless of the original distribution.

**Importance:**
- Allows us to use the normal distribution to make **inferences** about population parameters even if the population is not normal.
- Forms the basis of **confidence intervals and hypothesis tests**.

---

### **Q10: Assumptions of Central Limit Theorem**

1. **Samples are independent.**
2. **Sample size is large (n ≥ 30)**.
3. **Data is identically distributed (i.i.d).**
4. **Finite variance and mean**.
