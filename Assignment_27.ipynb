### **Q1. What is Statistics?**

**Statistics** is the branch of mathematics that deals with collecting, organizing, analyzing, interpreting, and presenting data to make informed decisions or draw meaningful conclusions.

---

### **Q2. Define the different types of statistics and give an example of when each type might be used.**

There are **two main types** of statistics:

1. **Descriptive Statistics**  
   - **Definition:** Summarizes or describes the main features of a dataset.
   - **Example:** Calculating the average score of students in a class.

2. **Inferential Statistics**  
   - **Definition:** Makes predictions or inferences about a population based on a sample.
   - **Example:** Estimating the percentage of voters who support a political party based on a poll.

---

### **Q3. What are the different types of data and how do they differ? Provide examples.**

Data can be classified as:

1. **Quantitative Data (Numerical):**
   - **Discrete:** Countable values  
     *Example:* Number of students in a class.
   - **Continuous:** Measurable values  
     *Example:* Height of students.

2. **Qualitative Data (Categorical):**
   - **Nominal:** Categories without order  
     *Example:* Blood groups (A, B, AB, O).
   - **Ordinal:** Categories with a logical order  
     *Example:* Satisfaction level (Low, Medium, High).

### **Q5. Explain the concept of levels of measurement and give an example of a variable for each level.**

1. **Nominal:** Categories with no order  
   *Example:* Gender (Male, Female)

2. **Ordinal:** Categories with a specific order  
   *Example:* Education level (High school, Bachelor's, Master's)

3. **Interval:** Numeric scales with equal intervals, no true zero  
   *Example:* Temperature in Celsius

4. **Ratio:** Numeric scales with equal intervals and a true zero  
   *Example:* Weight in kg

### **Q8. Which type of plot can be used to display data in terms of range?**

A **Box Plot** (Box-and-Whisker plot) is commonly used to display data ranges, showing minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

### **Q10. Common measures of central tendency and variability:**

**Central Tendency:**
1. **Mean (Average):** Sum of all values / Number of values  
   *Use:* Best for symmetrical distributions.
2. **Median:** Middle value  
   *Use:* Best for skewed distributions.
3. **Mode:** Most frequent value  
   *Use:* For categorical or unimodal data.

**Variability:**
1. **Range:** Difference between max and min  
   *Use:* Quick overview of spread.
2. **Variance:** Average squared deviation from mean  
   *Use:* Shows data dispersion.
3. **Standard Deviation:** Square root of variance  
   *Use:* Interpreted in the same units as data.
