### **Q1. What are the three measures of central tendency?**

1. **Mean** – The average of a dataset.  
2. **Median** – The middle value in a sorted dataset.  
3. **Mode** – The most frequently occurring value in a dataset.

---

### **Q2. What is the difference between mean, median, and mode?**

| Measure | Description | Use Case |
|--------|-------------|----------|
| **Mean** | Sum of values divided by number of values | Best for symmetrical data without outliers |
| **Median** | Middle value when data is sorted | Good for skewed data or data with outliers |
| **Mode** | Most frequent value(s) | Best for categorical data or unimodal distributions |

---

### **Q3. Calculate mean, median, and mode for:**

```python
import statistics as stats

data = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

mean = stats.mean(data)
median = stats.median(data)
mode = stats.mode(data)

mean, median, mode
```

**Answer:**
- **Mean ≈ 176.97**
- **Median = 177.0**
- **Mode = 177**

---

### **Q4. Find the standard deviation for the same data:**

```python
std_dev = stats.stdev(data)
std_dev
```

**Answer:**
- **Standard Deviation ≈ 1.85**

### **Q5. How are measures of dispersion used?**

- **Range:** Max - Min (Quick idea of spread)
- **Variance:** Average squared deviation from mean
- **Standard Deviation:** Spread of data around the mean

**Example:**
For data [10, 20, 30, 40, 50]:  
- Range = 40  
- Variance = 250  
- Standard deviation ≈ 15.81  
→ Higher std dev means more spread in data.

### **Q7. Sets A = {2,3,4,5,6,7}, B = {0,2,6,8,10}**

- **(i) A ∩ B (Intersection):** Elements common in A and B = **{2, 6}**
- **(ii) A ∪ B (Union):** All unique elements from A and B = **{0,2,3,4,5,6,7,8,10}**

### **Q8. What is skewness in data?**

**Skewness** measures the asymmetry of a distribution:

- **Right (positive) skew:** Long tail on the right
- **Left (negative) skew:** Long tail on the left
- **Zero skew:** Symmetric distribution

---

### **Q9. If data is right-skewed, where is the median relative to mean?**

**Right-skewed data:**  
- **Mean > Median**
- The mean is pulled in the direction of the long tail.

### **Q10. Difference between covariance and correlation:**

| Feature      | Covariance                             | Correlation                             |
|--------------|-----------------------------------------|------------------------------------------|
| Definition   | Measures how two variables change together | Standardized version of covariance       |
| Range        | Can be any value                       | Between -1 and +1                        |
| Unit         | Depends on variables                   | Unitless                                 |
| Use          | Detects direction of relation          | Measures strength & direction            |

### **Q11. Sample Mean Formula:**

\[
\bar{x} = \frac{\sum{x_i}}{n}
\]

**Example:**
For data [4, 5, 6]:
\[
\bar{x} = \frac{4+5+6}{3} = \frac{15}{3} = 5
\]

### **Q13. How is covariance different from correlation?**

- **Covariance:** Shows direction (positive/negative) but not strength.
- **Correlation:** Shows both **direction** and **strength** of the relationship on a standardized scale.

### **Q14. Effect of outliers on central tendency and dispersion:**

**Example:**  
Data = [10, 12, 13, 14, 500]

- **Mean = 109.8** (heavily affected by outlier)
- **Median = 13** (more robust)
- **Standard deviation** increases significantly due to outlier

**Conclusion:** Outliers distort the **mean** and **standard deviation**, but **median** remains relatively stable.
