### **Q1: What is Flask Framework? What are the advantages of Flask Framework?**
ANS:
Flask is a lightweight and flexible **Python web framework** used for building web applications. It is based on the **Werkzeug WSGI toolkit** and **Jinja2 template engine**, and it is categorized as a micro-framework because it does not require particular tools or libraries, allowing developers to have full control over the application.

#### **Advantages of Flask Framework:**
1. **Lightweight and Flexible:** Flask provides only the core features, making it simple and easy to customize.
2. **Ease of Use:** Its simple and straightforward syntax makes it beginner-friendly.
3. **Modular Design:** Supports extensions to add functionalities like authentication, database integration, etc.
4. **Built-in Development Server:** Includes a debugger and development server, enabling faster development.
5. **Scalability:** Suitable for small projects and can scale up to larger applications.
6. **Wide Community Support:** Large community ensures abundant resources, tutorials, and extensions.

### **Q2: Create a Simple Flask Application to Display ‘Hello World!!’**

#### Code:
```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return "Hello World!!"

if __name__ == '__main__':
    app.run(debug=True)
```

### **Q3: What is App Routing in Flask? Why Do We Use App Routes?**
ANS:
App routing refers to the process of mapping **URLs** to specific functions in a Flask application using the `@app.route` decorator. Each route corresponds to a specific view that processes and displays the content.

#### **Why Use App Routes?**
1. **URL Mapping:** Routes map user requests (URLs) to specific functions in the application.
2. **Customizable Endpoints:** Allows developers to define meaningful and user-friendly URLs.
3. **Multiple Views:** Enables creating multiple endpoints for different functionalities.
4. **Organized Application:** Keeps the application organized by associating each route with specific views.

### **Q4: Create a "/welcome" Route and a "/" Route**

#### Code:
```python
from flask import Flask

app = Flask(__name__)

# Route to display welcome message
@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

# Route to display company details
@app.route('/')
def company_details():
    details = """
    <h1>Company Details</h1>
    <p><strong>Company Name:</strong> ABC Corporation</p>
    <p><strong>Location:</strong> India</p>
    <p><strong>Contact Detail:</strong> 999-999-9999</p>
    """
    return details

if __name__ == '__main__':
    app.run(debug=True)
```

### **Q5: What Function is Used in Flask for URL Building?**
ANS:
The `url_for()` function is used in Flask to dynamically build URLs for specific functions.

#### **Benefits of `url_for()`:**
1. Automatically handles URL changes when routes are updated.
2. Simplifies the process of creating URLs in a Flask application.
3. Ensures consistency by linking URLs to function names.

#### **Python Code to Demonstrate `url_for()`:**
```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return f"Home Page - <a href='{url_for('welcome')}'>Go to Welcome Page</a>"

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

if __name__ == '__main__':
    app.run(debug=True)
```

#### Explanation:
- The `url_for('welcome')` dynamically generates the URL for the `welcome` route.
- Access the `/` route to see the link to the welcome page.

#### Output:
1. Visiting `/` displays:
   ```
   Home Page - Go to Welcome Page
   ```
2. Clicking the link navigates to `/welcome`.
