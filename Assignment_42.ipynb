Q1. Pearson Correlation – Study Time vs Exam Score

Pearson correlation coefficient measures the linear relationship between two variables.

Formula:
r = covariance(x, y) / (std(x) * std(y))

Example:

```python
from scipy.stats import pearsonr

study_time = [2, 3, 4, 5, 6]
exam_scores = [60, 65, 70, 75, 80]

r, p = pearsonr(study_time, exam_scores)
print("Pearson Correlation Coefficient:", r)
```

Interpretation:
A value close to 1 indicates a strong positive relationship.

Q2. Spearman Rank Correlation – Sleep vs Job Satisfaction

Spearman correlation measures the monotonic relationship between two variables.

```python
from scipy.stats import spearmanr

sleep_hours = [6, 7, 5, 8, 6]
job_satisfaction = [5, 6, 4, 9, 5]

rho, p = spearmanr(sleep_hours, job_satisfaction)
print("Spearman Rank Correlation:", rho)
```

Interpretation:
A value close to 1 indicates that as one variable increases, the other also increases.

Q3. Compare Pearson and Spearman – Exercise vs BMI

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

exercise_hours = np.random.randint(0, 10, 50)
bmi = np.random.normal(25, 3, 50)

pearson_corr, _ = pearsonr(exercise_hours, bmi)
spearman_corr, _ = spearmanr(exercise_hours, bmi)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```

Interpretation:
Compare both values. Pearson shows linear trend, Spearman shows overall trend direction.
