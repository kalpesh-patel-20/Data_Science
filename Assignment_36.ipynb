### **Q1: Explain the following with an example:**

#### **A) Artificial Intelligence (AI):**  
AI is the broader concept of machines being able to carry out tasks in a way that we would consider “smart.”  
**Example:** A self-driving car uses AI to detect objects, follow traffic rules, and make driving decisions.

#### **B) Machine Learning (ML):**  
ML is a subset of AI where machines learn from data to improve their performance without being explicitly programmed.  
**Example:** An email spam filter that learns from user behavior and patterns to detect spam messages.

#### **C) Deep Learning (DL):**  
DL is a subset of ML that uses neural networks with multiple layers (deep neural networks).  
**Example:** Face recognition systems use deep learning to accurately identify people in photos.

---

### **Q2: What is supervised learning? List some examples.**

**Supervised Learning** is a type of machine learning where the model is trained on labeled data (input-output pairs).  
It learns to map inputs to the correct output using the training data.

**Examples:**
- Predicting house prices (Regression)
- Email spam detection (Classification)
- Diagnosing diseases based on symptoms

---

### **Q3: What is unsupervised learning? List some examples.**

**Unsupervised Learning** deals with data that is not labeled. The model tries to identify hidden patterns or groupings in the data.

**Examples:**
- Customer segmentation
- Market basket analysis
- Dimensionality reduction (e.g., PCA)

---

### **Q4: What is the difference between AI, ML, DL, and DS?**

| Term | Description | Example |
|------|-------------|---------|
| **AI** | Broad field simulating human intelligence | Chatbots, Robotics |
| **ML** | Algorithms that learn from data | Spam detection |
| **DL** | Neural network-based ML | Image recognition |
| **DS** | Extraction of insights from data (includes AI/ML) | Data analysis dashboards |

### **Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?**

| Type | Data | Goal | Example |
|------|------|------|---------|
| **Supervised** | Labeled | Predict outcomes | Loan default prediction |
| **Unsupervised** | Unlabeled | Discover patterns | Customer segmentation |
| **Semi-supervised** | Small labeled + large unlabeled | Leverage both types | Image labeling with few annotated examples |

---

### **Q6: What is train, test, and validation split? Why is it important?**

- **Train Set**: Used to train the model.
- **Validation Set**: Used for tuning hyperparameters and preventing overfitting.
- **Test Set**: Used to evaluate final model performance.

**Importance**: Helps ensure the model generalizes well to unseen data.

---

### **Q7: How can unsupervised learning be used in anomaly detection?**

Unsupervised learning can detect data points that deviate significantly from the normal pattern or cluster.

**Example**:  
- In banking, clustering algorithms can identify unusual transactions as potential fraud.
- Isolation Forest or One-Class SVM can be used without labeled data.
